```python
# Function to process mseed files
def process_mseed_batch(file_paths):
    """
    Batch processes multiple .mseed files for efficiency.
    """
    streams = []
    for file_path in file_paths:
        try:
            # Read the mseed file
            stream = obspy.read(file_path)
            streams.append(stream)
            # Print some basic information
            print(f"File processed successfully: {file_path}")
        except Exception as e:
            # Error handling for missing or corrupted files
            print(f"Error reading {file_path}: {e}")
    return streams

all_streams = process_mseed_batch(file_paths)
```

```python
import obspy
import matplotlib.pyplot as plt
import numpy as np

# Preprocess function
def preprocess_stream(stream, filter_min_freq, filter_max_freq, taper_pct=0.05):
    try:
        for trace in stream:
            # Basic preprocessing steps with error handling
            trace.detrend(type='linear')
            trace.taper(max_percentage=taper_pct, type='cosine')
            trace.filter('bandpass', freqmin=filter_min_freq, freqmax=filter_max_freq)
        return stream
    except Exception as e:
        print(f"Error during preprocessing: {e}")
        return None  # Return None if preprocessing fails

# FFT Plotting function to visualize frequency content
def plot_fft(trace):
    try:
        npts = trace.stats.npts
        sampling_rate = trace.stats.sampling_rate
        freqs = np.fft.fftfreq(npts, d=1 / sampling_rate)
        fft_values = np.fft.fft(trace.data)

        plt.figure(figsize=(10, 6))
        plt.plot(np.abs(freqs), np.abs(fft_values))
        plt.title("FFT of the Seismic Data")
        plt.xlabel("Frequency (Hz)")
        plt.ylabel("Amplitude")
        plt.xlim(0, 0.005)
        plt.show()
    except Exception as e:
        print(f"Error in FFT plotting: {e}")

# Define filter frequency range
filter_min_freq = 0.001  # Minimum frequency (Hz)
filter_max_freq = 0.004  # Maximum frequency (Hz)

preprocessed_streams = []
for stream in all_streams:
    if stream is not None:
        preprocessed_stream = preprocess_stream(stream, filter_min_freq, filter_max_freq, taper_pct=0.03)
        if preprocessed_stream is not None:
            preprocessed_streams.append(preprocessed_stream)


for stream in preprocessed_streams:
    try:
        # Plot original and preprocessed data
        stream.plot()
        print("Original Data FFT:")
        plot_fft(stream[0])
    except Exception as e:
        print(f"Error during plotting or FFT analysis: {e}")
```

```python
from scipy.fftpack import fft
from scipy.stats import skew, kurtosis
import numpy as np
from scipy.stats import skew, kurtosis, entropy  # Added entropy import


def get_label_for_trace(trace):
    try:
        threshold = 1000  # Define a threshold for amplitude
        if np.max(trace.data) > threshold:
            return 1  # Event detected (label as 1)
        else:
            return 0  # No event detected (label as 0)
    except Exception as e:
        print(f"Error in labeling trace: {e}")
        return None
    
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Updated extract_features function
# Feature Extraction Function (Updated)
def extract_features(trace, window_size=50, step_size=5):
    feature_matrix = []
    data = trace.data

    if len(data) == 0:
        print("Empty trace data, skipping...")
        return np.array([])

    try:
        # Calculate the number of windows
        num_windows = (len(data) - window_size) // step_size + 1
        print(f"Number of windows: {num_windows}")

        for start in range(0, len(data) - window_size + 1, step_size):
            end = start + window_size
            window = data[start:end]

            if len(window) < window_size:
                print(f"Skipping window due to insufficient data: {len(window)}")
                continue

            # Existing and new feature calculations
            max_amp = np.max(window)
            min_amp = np.min(window)
            mean_amp = np.mean(window)
            skewness = skew(window, bias=False)
            kurt = kurtosis(window, bias=False)

            # Frequency domain features
            freq_data = np.abs(fft(window))
            mean_freq = np.mean(freq_data)
            max_freq = np.max(freq_data)
            peak_freq = np.argmax(freq_data)
            freq_variance = np.var(freq_data)

            # Additional Features
            zero_crossings = np.where(np.diff(np.sign(window)))[0]
            zero_crossing_rate = len(zero_crossings) / len(window)

            power_spectrum = np.abs(freq_data) ** 2
            power_spectrum = power_spectrum / np.sum(power_spectrum)  # Normalize
            spectral_entropy = entropy(power_spectrum)

            # Updated feature vector
            feature_vector = [
                max_amp, min_amp, mean_amp, skewness, kurt,
                mean_freq, max_freq, peak_freq, freq_variance,
                zero_crossing_rate, spectral_entropy
            ]
            feature_matrix.append(feature_vector)

    except Exception as e:
        print(f"Error during feature extraction: {e}")

    return np.array(feature_matrix)

# Process and extract features for all streams
all_features = []
all_labels = []

for stream in preprocessed_streams:
    for trace in stream:
        features = extract_features(trace, window_size=50, step_size=5)
        label = get_label_for_trace(trace)

        if features.size > 0 and label is not None:
            all_features.extend(features)
            all_labels.extend([label] * len(features))

# Convert to NumPy arrays and ensure they have the correct shape
all_features = np.array(all_features)
all_labels = np.array(all_labels)

if all_features.size == 0:
    print("Error: No features extracted. Please check the data and extraction process.")
else:
    print(f"Features shape: {all_features.shape}")
    print(f"Labels shape: {all_labels.shape}")

    # Ensure features are in 2D shape
    if len(all_features.shape) == 1:
        all_features = all_features.reshape(-1, 1)

    # **Standardization of Features**
    scaler = StandardScaler()
    all_features = scaler.fit_transform(all_features)

    # **Split Data for Training and Testing**
    X_train, X_test, y_train, y_test = train_test_split(all_features, all_labels, test_size=0.3, random_state=42)

    # **Verify the shapes of the split data**
    print(f"Training features shape: {X_train.shape}")
    print(f"Test features shape: {X_test.shape}")
    print(f"Training labels shape: {y_train.shape}")
    print(f"Test labels shape: {y_test.shape}")

    # **Train a Random Forest Classifier**
    clf = RandomForestClassifier(n_estimators=100, random_state=42)
    clf.fit(X_train, y_train)

    # **Predict on Test Data**
    y_pred = clf.predict(X_test)

    # **Evaluate the Model**
    print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
    print(classification_report(y_test, y_pred))
```
