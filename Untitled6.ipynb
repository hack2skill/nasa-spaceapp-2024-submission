```python

def preprocess_stream(stream, filter_freq):
    # Ensure filter_freq is below Nyquist frequency
    for trace in stream:
        sampling_rate = trace.stats.sampling_rate
        nyquist = sampling_rate / 2.0
        if filter_freq >= nyquist:
            raise ValueError(f"Filter frequency {filter_freq} is too high. The Nyquist frequency is {nyquist}. Please lower the filter frequency.")

        # Detrend, taper, and filter the trace
        trace.detrend(type='linear')
        trace.taper(max_percentage=0.05, type='cosine')
        trace.filter('lowpass', freq=filter_freq)

    return stream

# Define the filter frequency (make sure it's appropriate for your data)
filter_freq = 0.001  # Example value; adjust based on your data

for file in file_paths:
    stream = obspy.read(file)
    preprocessed_stream = preprocess_stream(stream, filter_freq)
    preprocessed_stream.plot()  # Optional, plot the preprocessed data
```

```python
from obspy.signal.trigger import classic_sta_lta, trigger_onset

def detect_events(trace, sta_length_factor=0.1, lta_length_factor=1.0, trigger_on=1.5, trigger_off=0.5):
    data_length = len(trace.data)

    # Determine sta_length and lta_length based on data length
    sta_length = max(int(data_length * sta_length_factor), 1)
    lta_length = max(int(data_length * lta_length_factor), sta_length + 1)

    if data_length < lta_length:
        print(f"Data length ({data_length}) is too short for LTA window size ({lta_length}) in trace: {trace.stats.network}.{trace.stats.station}")
        return []

    # Apply STA/LTA algorithm for event detection
    sta_lta = classic_sta_lta(trace.data, nsta=sta_length, nlta=lta_length)

    # Define trigger thresholds for detection
    triggers = trigger_onset(sta_lta, trigger_on, trigger_off)

    # Filter out duplicate events
    unique_triggers = []
    for trigger in triggers:
        if trigger.tolist() not in unique_triggers:
            unique_triggers.append(trigger.tolist())

    if len(unique_triggers) > 0:
        print(f"Detected {len(unique_triggers)} potential events in trace: {trace.stats.network}.{trace.stats.station}")
        return unique_triggers
    else:
        print("No significant events detected.")
        return []

def plot_events(trace, events):
    plt.figure(figsize=(10, 6))
    plt.plot(trace.times(), trace.data, 'k', label='Trace Data')
    for event in events:
        plt.axvline(x=trace.times()[event[0]], color='r', linestyle='--', label='Detected Event')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.title(f'Event Detection for {trace.stats.network}.{trace.stats.station}')
    plt.legend()
    plt.show()


filter_freq = 0.001  # Example filter frequency; adjust as needed

for file in file_paths:
    stream = obspy.read(file)
    preprocessed_stream = preprocess_stream(stream, filter_freq)

    for trace in preprocessed_stream:
        detected_events = detect_events(trace)
        if len(detected_events) > 0:
            print(f"Detected events: {detected_events}")
            plot_events(trace, detected_events)
        else:
            print("No events detected in this trace.")
```

```python
import numpy as np
from obspy.signal.trigger import classic_sta_lta, trigger_onset

def label_data(trace, sta_length=1.0, lta_length=10.0, threshold_on=2.5, threshold_off=1.5):
    # Apply STA/LTA to detect events
    sta_lta = classic_sta_lta(trace.data, nsta=int(sta_length * trace.stats.sampling_rate),
                              nlta=int(lta_length * trace.stats.sampling_rate))
    triggers = trigger_onset(sta_lta, threshold_on, threshold_off)

    # Label the data as event or noise
    labels = np.zeros(len(trace.data))
    for trigger in triggers:
        labels[trigger[0]:trigger[1]] = 1  # Label event regions as 1 (event), others as 0 (noise)

    return labels

# Example usage:
for trace in preprocessed_stream:
    labels = label_data(trace)

```

```python
from scipy.fftpack import fft
from scipy.stats import skew, kurtosis
import numpy as np

def extract_features(trace, window_size=75):
    # Split trace data into windows and extract features
    feature_matrix = []  # To store the feature vectors for all windows
    data = trace.data

    # Check if data is empty
    if len(data) == 0:
        print("Empty trace data, skipping...")
        return np.array([])  # Return an empty array if no data

    num_windows = len(data) // window_size  # Number of windows to process
    print(f"Number of windows: {num_windows}")

    for i in range(num_windows):
        window = data[i*window_size:(i+1)*window_size]

        # Skip windows with insufficient data
        if len(window) < window_size:
            print(f"Skipping window {i} due to insufficient data: {len(window)}")
            continue

        # Amplitude-based features
        max_amp = np.max(window)
        min_amp = np.min(window)
        mean_amp = np.mean(window)
        skewness = skew(window)
        kurt = kurtosis(window)

        # Frequency-based features (using FFT)
        freq_data = np.abs(fft(window))
        mean_freq = np.mean(freq_data)
        max_freq = np.max(freq_data)

        # Create a feature vector for the current window
        feature_vector = [max_amp, min_amp, mean_amp, skewness, kurt, mean_freq, max_freq]
        feature_matrix.append(feature_vector)  # Append the feature vector for this window

    return np.array(feature_matrix)  # Return the feature matrix for all windows

import numpy as np

# Initialize empty lists to hold features and labels
all_features = []
all_labels = []

# Process each trace to extract features
for trace in preprocessed_stream:
    features = extract_features(trace)

    # For this example, assuming you have a label for each trace
    # Adjust label extraction based on your data and labeling strategy
    label = get_label_for_trace(trace)  # Replace this with actual label extraction

    if features.size > 0 and label is not None:
        all_features.extend(features)
        all_labels.extend([label] * len(features))  # Repeat label for each feature window

# Convert lists to numpy arrays
all_features = np.array(all_features)
all_labels = np.array(all_labels)

print(f"Features shape: {all_features.shape}")
print(f"Labels shape: {all_labels.shape}")

# Now, you should have matching feature and label sizes
```
